```python

# Import pandas and read data into a data frame
import pandas as pd
df = pd.read_csv("data/breast-cancer.csv")

df.info()
```

```python
# There are no null values :)
df.isnull().sum()
```

```python
df.head(196)
```

```python
# We can channge diagnosis to be a 0,1 False, True
# and we can drop the id
df["diagnosis"].replace({"M": 1}, inplace=True)
df["diagnosis"].replace({"B": 0}, inplace=True)
df = df.drop("id" , axis=1)
df.head(10)
```

```python
# Now that our data is clean we can start the decision tree proces
# First split the data into X and y. Where X will be the input data
# and y will be what we want to predict which is Survived
from sklearn.model_selection import train_test_split
X = df.iloc[:, 1:31]
y = df.iloc[:, 0]

# Split Data into test and training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

print(X_test)
#y.head()
```

```python
# .getparams() To see what changes you can do 
```

```python
# Now we can create the Decision Tree Classifier and fit out training data to the modle
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train, y_train)
```

```python
# Now we can test 
predictions = dtc.predict(X_test)


# Returns an array of 0,1 where 0 is that they dont have cancer, 1 is they have cancer
predictions 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
# https://en.wikipedia.org/wiki/F-score
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
# top left -> True Positive
# top right -> False Negative
# bottom left -> False Positive
# bottom right -> True Negative
# https://en.wikipedia.org/wiki/F-score
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```
